## Practice Assignment for all functions of String

### String - capitalize, casefold, center, count, encode, endswith, expandtab, find, format, format_map, index

```python
my_str = "hello world 123"
print("original string =", my_str)

print("capitalize =", my_str.capitalize())
print("casefold =", my_str.casefold())
print("center (width 20) =", my_str.center(20))
print("count of 'l' =", my_str.count('l'))
print("encode =", my_str.encode())
print("endswith '123' =", my_str.endswith('123'))
print("expandtab =", "hello\tworld".expandtabs(4))
print("find 'world' =", my_str.find('world'))
print("format =", "Hello, {}".format("Alice"))
print("format_map =", "Hello, {name}".format_map({'name': 'Bob'}))
print("index of 'world' =", my_str.index('world'))
```

### String  - isalnum, isalpha, isascii, isdecimal, isdigit, isidentifier, islower, isnumeric, isprintable, isspace, istitle, isupper

```python
my_str = "hello123"
print("original string =", my_str)

print("isalnum =", my_str.isalnum())
print("isalpha =", my_str.isalpha())
print("isascii =", my_str.isascii())
print("isdecimal =", "12345".isdecimal())
print("isdigit =", "12345".isdigit())
print("isidentifier =", "my_var".isidentifier())
print("islower =", my_str.islower())
print("isnumeric =", "12345".isnumeric())
print("isprintable =", my_str.isprintable())
print("isspace =", "   ".isspace())
print("istitle =", my_str.istitle())
print("isupper =", my_str.isupper())
```

### String  -join, ljoin, lower, lstrip, maketeans, partition, replace, rfind, rindex, rjust, rpartition, rstrip, split, splitlines, startswith, strip, swapcase, title, trenslate, upper, zfill

```python
my_str = " hello world "
print("original string =", my_str)

print("join =", ", ".join(['apple', 'banana', 'cherry']))
print("lower =", my_str.lower())
print("lstrip =", my_str.lstrip())
print("maketrans =", my_str.translate(str.maketrans("aeiou", "12345")))
print("partition =", my_str.partition('world'))
print("replace 'world' with 'everyone' =", my_str.replace("world", "everyone"))
print("rfind 'world' =", my_str.rfind("world"))
print("rindex 'world' =", my_str.rindex("world"))
print("rjust (width 20) =", my_str.rjust(20))
print("rpartition =", my_str.rpartition(' '))
print("rstrip =", my_str.rstrip())
print("split =", my_str.split())
print("splitlines =", "hello\nworld\npython".splitlines())
print("startswith ' hello' =", my_str.startswith(" hello"))
print("strip =", my_str.strip())
print("swapcase =", my_str.swapcase())
print("title =", my_str.title())
print("translate =", my_str.translate(str.maketrans("aeiou", "12345")))
print("upper =", my_str.upper())
print("zfill =", "42".zfill(5))
```
