###  Dictionary - pop,popitem,copy,clear,update,key,values,items,get

```python
my_dict={'a':1,'b':2,'k':"kumar"}
print("type of my_dict=",type(my_dict))
print("original dictionary=",my_dict)

pop_1=my_dict.pop('b')
print("popped item=",pop_1)
print("after popping b from the dict=",my_dict)

popitem_1=my_dict.popitem()
print("popped item=", popitem_1)
print("popitem remove the last key-value pair=",my_dict)

new_dict = my_dict.copy()
print("copy of the dict=",new_dict) 

my_dict.clear()
print("my_dict after using clear=",my_dict)
print("new_dict is=",new_dict)


my_dict.update({'h':8})
my_dict.update({3:'t'})
print("after adding two key-value pair to my dict=",my_dict)

keys=my_dict.keys()
print("keys in dict=",keys)

values=my_dict.values()
print("values in dict=",values)

items=my_dict.items()
print("items in dict=",items)

value_1=my_dict.get('h',)
value_2=my_dict.get('t',)
value_3=my_dict.get(3,'default_value')
print("the value of h=",value_1)
print("the value of t=",value_2)
print("the value of 3=",value_3)
```

### Dictionary -fromkeys,setdefault,contains,delitem,getitem

```python
my_tuple = (1, 3, 5, 2, 4, 7, 5, 3, 25, 36, 47, 59, 60)
print("original tuple =", my_tuple)

print("length of tuple =", len(my_tuple))
print("smallest element of tuple =", min(my_tuple))
print("largest element of tuple =", max(my_tuple))
print("sum of all elements of tuple =", sum(my_tuple))

my_tuple2 = (10, 20, 30)
print("tuple 2=",my_tuple2)
print("concatenated tuple =", my_tuple + my_tuple2)
print("repeated tuple =", my_tuple * 2)
print("tuple after slicing (2:6) =", my_tuple[2:6])

nested_tuple = (1, (2, 3), 4)
print("accessing nested tuple =", nested_tuple[1][1])  
```
