## Practice Assignment for all functions of List

### list -append,extend,insert,remove,pop,clear

```python
my_list=[2,2,34,"ashish",8.5,'a']
print("original list=", my_list)

my_list.append("kumar")
my_list.append(5)
print("append - kumar,5 =",my_list)

my_list.extend(['k',6,"yadav"])
print("extend - k,6,yadav =",my_list)

my_list.insert(1,1)
my_list.insert(-1,4)
print("insert - 1,4 at index 1,-1 resp. =",my_list)

my_list.remove(2)
print("remove - 2 from my list(first occurrence) =",my_list)

my_list.pop(2)
my_list.pop(1)
print("pop the element at the index 2,1 =",my_list)

my_list.clear()
print("clear all the elements of list=",my_list)
```

### list -index,count,sort,reverse,copy

```python
my_list=[1,7,4,89,43,0,7,4,7]
print("original list=", my_list)

index_1=my_list.index(89)
index_2=my_list.index(0)
print("index of the element 89 and 0 is ",index_1, "and" ,index_2)

count_1=my_list.count(7)
count_2=my_list.count(4)
print("count occurrences of 7 and 4 is ",count_1, "and" ,count_2)

my_list.sort()
print("sort() for ascending order  =",my_list)
my_list.sort(reverse=False)
print("sort(reverse=False) for ascending order =",my_list)
my_list.sort(reverse=True)
print("sort(reverse=True) for descending order =",my_list)

my_list.reverse()
print("reverse the order of list = ",my_list)

new_list = my_list.copy()
new_list.append(6)
print("Original List:", my_list)  
print("Copied List:", new_list)   

```

###  list -len,min,max,sum

```python
my_list=[1,3,5,2,4,7,5,3,25,36,47,59,60]
print("original list=", my_list)

print("length of list =",len(my_list))
print("smallest element of list=",min(my_list))
print("largest element of list=",max(my_list))
print("sum of all element of list=",sum(my_list))
```
