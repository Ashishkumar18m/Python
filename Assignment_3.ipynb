### Q14. Count Vowels and Consonants in a String

```python
my_string = "I am writing python code in jupyter"
vowels = "aeiouAEIOU"
vowel_count = 0
consonant_count = 0

for char in my_string:
    if char.isalpha():  
        if char in vowels:
            vowel_count += 1
        else:
            consonant_count += 1

print("Vowels:", vowel_count)
print("Consonants:", consonant_count)
```

### Q17. Find the Frequency of Each Word in a Tuple of Sentences

```python
sentences_tuple = ("hello world", "hello there", "world of python", "python world")

word_count = {}

for sentence in sentences_tuple:
    words = sentence.split()
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1

print("Word frequency:", word_count)
```

### Q20. Reverse a Tuple Using a For Loop

```python
my_tuple = (1, 2, 3, 4, 5)
reversed_tuple = ()

for element in my_tuple[::-1]:  
    reversed_tuple += (element,)  

print("Reversed Tuple:", reversed_tuple)
```
