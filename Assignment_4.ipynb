### 1. Write a function that takes a list of numbers and returns the sum of all elements using a for loop.

```python
def sum_of_elements(my_list):
    total=0
    for num in my_list:
        total+=num

    return total

my_list=[1,2,3,4,5,6]
result= sum_of_elements(my_list)
print("sum of elements=",result)
```

### 2. Write a function that removes all duplicate elements from a list using a while loop.

```python
def remove_duplicates(my_list):
    index = 0
    while index < len(my_list):

        if my_list[index] in my_list[index+1:]:
            my_list.remove(my_list[index])
        else:
            index += 1
    return my_list
my_list=[1,2,3,5,2,7,3,5,68,3]
result=remove_duplicates(my_list)
print("unique list=",result)
```

### 3. Create a function that reverses a list without using built-in functions.

```python
def reverse_list(input_list):
    reversed_list = [] 
    for i in range(len(input_list)-1,-1,-1 ):
        reversed_list.append(input_list[i])
    return reversed_list


my_list = [1, 2, 3, 4, 5] 
reversed_list = reverse_list(my_list)  
print("Original list:", my_list) 
print("Reversed list:", reversed_list)  
```

### 5. Implement a function that checks if a given element exists in a list using a while loop.

```python
def element_exists(input_list, element):
    index = 0 
    while index < len(input_list): 
        if input_list[index] == element: 
            return True  
        index += 1  
    return False  


my_list = [10, 20, 30, 40, 50]
element_to_check = 30 

exists = element_exists(my_list, element_to_check)

if exists:
    print(f"Element {element_to_check} exists in the list.")
else:
    print(f"Element {element_to_check} does not exist in the list.")
```

### 6. Write a function that takes a tuple of numbers and returns a new tuple with only even numbers.

### 7. Write a function that finds the index of a given element in a tuple using a for loop.

```python
def find_element_index(input_tuple, element):
   
    for index, current_element in enumerate(input_tuple):
        if current_element == element:
            return index  
    
    return -1


numbers_tuple = (10, 20, 30, 40, 50)
result = find_element_index(numbers_tuple, 30)
print("Index of the element:", result)

result_not_found = find_element_index(numbers_tuple, 60)
print("Index of the element:", result_not_found)
```

### 8. Create a function that converts a tuple of numbers into a list, modifies it, and returns the new tuple.

```python
def modify_tuple(input_tuple):
    temp_list = list(input_tuple)
    
    for i in range(len(temp_list)):
        temp_list[i] += 1
    
    modified_tuple = tuple(temp_list)
    return modified_tuple

numbers_tuple = (1, 2, 3, 4)
result = modify_tuple(numbers_tuple)
print("Modified tuple:", result)
```

### 9. Implement a function to merge two tuples using a for loop.

```python
def merge_tuples(tuple1, tuple2):
    merged_tuple = ()
    for item in tuple1:
        merged_tuple += (item,) 
    for item in tuple2:
        merged_tuple += (item,)  
    return merged_tuple

tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
result = merge_tuples(tuple1, tuple2)
print(result)
```

### 12. Create a function that finds the union of two sets using a for loop.

```python
def union_of_sets(set1, set2):
    
    union_set = set()
    for num in set1:
        union_set.add(num)

    for num in set2:
        union_set.add(num)
    
    return union_set


set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

result = union_of_sets(set1, set2)
print("Union of the sets:", result)
```

### 13. Write a function to check if all elements of one set exist in another set using a loop.

```python
def is_subset(set1, set2):
   
    for num in set1:
       
        if num not in set2:
            return False
   
    return True


set1 = {1, 2, 3}
set2 = {1, 2, 3, 4, 5}

result = is_subset(set1, set2)
print("All elements of set1 exist in set2:", result)
```

### 14. Implement a function that finds the difference between two sets without using built-in methods.

```python
def difference_of_sets(set1, set2):
    diff_set = set()
    
   
    for num in set1:
        
        if num not in set2:
            diff_set.add(num)
    
    return diff_set


set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

result = difference_of_sets(set1, set2)
print("Difference between set1 and set2:", result)
```

### 15. Write a function that converts a set to a list and returns the sorted list using a for loop.

```python
def set_to_sorted_list(input_set):
    
    input_list = list(input_set)
    
    
    for i in range(len(input_list)):
        for j in range(len(input_list) - i - 1):
            if input_list[j] > input_list[j + 1]:
                
                input_list[j], input_list[j + 1] = input_list[j + 1], input_list[j]
    
    return input_list


input_set = {5, 3, 8, 1, 2}
sorted_list = set_to_sorted_list(input_set)
print("Sorted list:", sorted_list)
```

### 16. Write a function that takes a dictionary and returns a list of all its keys using a loop.

```python
def get_keys_from_dict(input_dict):
    
    keys_list = []
    
    
    for key in input_dict:
        keys_list.append(key)
    
    return keys_list


sample_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
keys = get_keys_from_dict(sample_dict)
print("Keys in the dictionary:", keys)
```

### 17. Create a function that finds the key with the highest value in a dictionary.

```python
def key_with_highest_value(input_dict):
   
    if not input_dict:
        return None
    
    
    highest_key = None
    highest_value = float('-inf') 
    
    for key, value in input_dict.items():
        if value > highest_value:
            highest_value = value
            highest_key = key
    
    return highest_key


sample_dict = {'a': 10, 'b': 20, 'c': 30, 'd': 5}
result = key_with_highest_value(sample_dict)
print("Key with the highest value:", result)
```

### 18. Write a function that reverses keys and values in a dictionary using a for loop.

```python
def reverse_dict(input_dict):
    reversed_dict = {}
    
    for key, value in input_dict.items():
        reversed_dict[value] = key
    
    return reversed_dict

sample_dict = {'a': 1, 'b': 2, 'c': 3}
reversed_result = reverse_dict(sample_dict)
print("Reversed dictionary:", reversed_result)
```

### 19. Implement a function that counts occurrences of each word in a list and stores them in a dictionary.

```python
def count_word_occurrences(word_list):
    word_count = {}
    
    for word in word_list:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    
    return word_count

words = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
result = count_word_occurrences(words)
print("Word occurrences:", result)
```

### 20. Write a function that removes dictionary entries where values are less than a given threshold using a while loop.

```python
def remove_entries_below_threshold(input_dict, threshold):
    items = list(input_dict.items())
    
    index = 0
    
    while index < len(items):
        key, value = items[index]
        
        if value < threshold:
            del input_dict[key]
        
        index += 1
    
    return input_dict

sample_dict = {'a': 10, 'b': 2, 'c': 5, 'd': 1}
threshold = 5
result = remove_entries_below_threshold(sample_dict, threshold)
print("Modified dictionary:", result)
```
