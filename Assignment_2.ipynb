## 13. Given a list of numbers, use an if condition inside a loop to remove all even numbers.

```python
list1 = input("Enter numbers: ").split() 
list2 = []

i = 0
while i < len(list1):
    if int(list1[i]) % 2 == 0:  
        list2.append(list1[i])
    i += 1  

print("Original list:", list1)
print("List of even numbers:", list2)
```

## 15. Write a function to check if a given string is a palindrome (reads the same forward and backward). If it is, print "Palindrome"; otherwise, print "Not a palindrome".

```python
def check(s1):
    s2 = s1[:: -1]
    print("Palindrome" if (s1 == s2) else "Not a Palindrome")

s1 = input("enter: ")
check(s1)
```

## 16. Given a string, count the number of vowels (a, e, i, o, u). If the count is greater than 5, print "Lots of vowels", else print "Few vowels".

```python
s = 'aslam walekum'

count = sum(1 for char in s.lower() if char in "aeiou")
print("Lots of vowels" if count > 5 else "Few vowels")
```

## 19. Write a function that checks the strength of a password based on these conditions:
## • If the length is less than 6, print "Weak password".
## • If it contains only letters, print "Moderate password".
## • If it contains letters and numbers, print "Strong password".

```python
def strength(password):
    if len(password)<6 :
        print("Weak Password")
    elif password.isalpha():
        print("Moderate Password")
    elif password.isalnum():
        print("Strong  Password")

p = input("enter password: ")
strength(p)

```
