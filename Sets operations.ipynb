### Sets -add,remove,pop,discard,clear,len

```python
my_set={1,2,3,"ashish",45,78,6.5}
print(type(my_set))
print("original set=", my_set)

my_set.add(4)
my_set.add("kumar")
print("after adding 4 , kumar in set =",my_set)

print("the length of set=",len(my_set))

remove_1=my_set.remove(2)
remove_2=my_set.remove("ashish")
print("after removing 2,ashish=",my_set)

popped_item_1=my_set.pop()
popped_item_2=my_set.pop()
print("popped item=",popped_item_1,popped_item_2)
print("after popping 2 element at random=",my_set)

my_set.discard(4)
my_set.discard("kumar")
print("after discarding 4,kumar =",my_set)

my_set.clear()
print("clear-clears the entire set=",my_set)

```

### Sets - union,intersection,difference,symmetric_difference,isdisjoint

```python
my_set1={1,2,3,5,6,7}
my_set2={2,5,8,9,4,7}
print("original set1 and set2 are=", my_set1,my_set2)

union_set1 = my_set1.union(my_set2)  
union_set2 = my_set1 | my_set2 
print("union of set 1,2 using union =",union_set1)
print("union of set 1,2 using |=",union_set2)

intersection_set1 = my_set1.intersection(my_set2)  
intersection_set2 = my_set1 & my_set2 
print("intersection of set 1,2 using intersection =",intersection_set1)
print("intersection of set 1,2 using &=",intersection_set2)

difference_set1 = my_set1.difference(my_set2)  
difference_set2 = my_set1 - my_set2 
print("difference of set 1,2 using difference =",difference_set1)
print("difference of set 1,2 using - =",difference_set2)

symmetric_difference_set1 = my_set1.symmetric_difference(my_set2)  
symmetric_difference_set2 = my_set1 ^ my_set2 
print("symmetric_difference of set 1,2 using  symmetric_difference =", symmetric_difference_set1)
print("symmetric_difference of set 1,2 using ^ =", symmetric_difference_set2)

print("return True if the sets have no elements is common else False =",my_set1.isdisjoint(my_set2) )


```

### Sets  -copy,difference,difference update

```python
my_set = {1, 2, 3, 4, 5}
other_set = {4, 5, 6, 7, 8}

print("original set =", my_set)
print("other set =", other_set)

copied_set = my_set.copy()
print("copied set =", copied_set)

difference_set = my_set.difference(other_set)
print("difference of my_set and other_set =", difference_set)

my_set.difference_update(other_set)
print("my_set after difference_update with other_set =", my_set)
```

### Set  - intersection_update,isdisjoint,issubset,issuperset

```python
my_set = {1, 2, 3, 4, 5}
other_set = {4, 5, 6, 7, 8}

print("original set =", my_set)
print("other set =", other_set)

my_set = {1, 2, 3, 4, 5}
my_set.intersection_update(other_set)
print("my_set after intersection_update with other_set =", my_set)

print("is my_set disjoint with other_set?", my_set.isdisjoint(other_set))

print("is my_set a subset of other_set?", my_set.issubset(other_set))

print("is my_set a superset of other_set?", my_set.issuperset(other_set))
```

### Set  -symmetric_difference_update,update

```python
my_set = {1, 2, 3, 4, 5}
other_set = {4, 5, 6, 7, 8}

print("original set =", my_set)
print("other set =", other_set)

my_set = {1, 2, 3, 4, 5}
my_set.symmetric_difference_update(other_set)
print("my_set after symmetric_difference_update with other_set =", my_set)

my_set = {1, 2, 3}
my_set.update(other_set)
print("my_set after update with other_set =", my_set)
```
