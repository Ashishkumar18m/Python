#### 34. Write a program to merge two lists and remove duplicates.

```python
list1  = [324,12,32,45,87,3,6,76,7]


list2  = [12,32,45,65,87,87,68,7,674,2,56,76,67]
merged = list1 + list2

i = 0
while i < len(merged):
    j= i+1
    while j<len(merged):
        if merged[i] == merged[j]:
            merged.pop(j)
           
        j+=1
    i+=1

print("The final merged list after eliminating duplicates: ",merged)
```
